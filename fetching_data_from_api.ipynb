```python
import pandas as pd
import requests
```

```python
# taking the data from api by sending request
data=requests.get("https://api.nasa.gov/techtransfer/patent/?engine&api_key=YOURAPIKEY")
```

```python
# so basically here we are doing two things the first preprocessing the data and only taking the coloumns [[0,1,3,5,12]] and taking only results. Also converting the data frame to json
df=pd.DataFrame(data.json()['results'])[[0,1,3,5,12]]
df.head()
```

```python
# setting the id as index
df.set_index('id',inplace=True)
df
```

```python
df_temp=pd.DataFrame()
```

```python
# when you are given with large data sets then you can do this to create a data frame at once
for i in range(1,37):
  data=requests.get("https://api.nasa.gov/techtransfer/patent/?engine&api_key=80CFFM13x8JfYKhXcPvQlHlKtqT7qbz0C8nwdoih")
  df=pd.DataFrame(data.json()['results'])[[0,1,3,5,12]]
  df_temp=pd.concat([df_temp, df], ignore_index=True)
```

```python
df_temp
```

```python
# Naming the coloumns of the data sets
df_temp.columns=['id','Engine-Model','Overview','Usage','Ratings']
df_temp
```

```python
# checking for duplicates
duplicates=df_temp.duplicated()
if duplicates.any():
  print("It consist of duplicates values")
```

```python
# Copying the values into a new data frame
tech_transfer=df_temp.head(36).copy()
tech_transfer.head()
```

```python
# converting the data into a csv data sets
tech_transfer.to_csv("Nasa-tech-transfer.csv")
```
