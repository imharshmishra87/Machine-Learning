```python
tips=sns.load_dataset('tips')
tips.head()
```

# Scatter Plot (Numerical-Numerical)

```python
sns.relplot(data=tips, x='total_bill',y='tip',kind='scatter',hue='sex',style='smoker',size='size')
```

# Bar Plot(Numerical-Categorical)

```python
titanic=pd.read_csv("/content/Titanic-Dataset.csv")
titanic.head()
```

```python
sns.barplot(data=titanic, x='Pclass',y='Fare',hue='Sex')
```

```python
sns.barplot(data=titanic,x='Embarked',y='Age',hue='Sex')
```

# Box Plot(Numerical-Categorical)

```python
sns.boxplot(data=titanic,x='Sex',y='Age',hue='Survived')
```

# Distplot(Numerical-Categorical)

```python
sns.distplot(titanic[titanic['Survived']==0]['Fare'],hist=False)
sns.distplot(titanic[titanic['Survived']==1]['Fare'],hist=False)
```

# Heat Map(Categorical-categorical)

```python
titanic.head(3)
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

```python
(titanic.groupby('Pclass').mean(numeric_only=True)['Survived']*100).plot(kind='bar',ylabel='Survived percentage')
```

# Cluster Map(Categorical-categorical)

```python
sns.clustermap(pd.crosstab(titanic['Parch'],titanic['Survived']))
```

# Pairplot

```python
iris=sns.load_dataset('iris')
iris.head()
```

```python
sns.pairplot(data=iris,hue='species')
```

# Lineplot(Numerical - Numerical)

```python
flight=sns.load_dataset('flights')
flight.head()
```

```python
new=flight.groupby('year',observed=True).sum(numeric_only=True)
pd.DataFrame(new)
new.head()
```

```python
sns.lineplot(data=new,x='year',y='passengers')
```

```python
flight.pivot_table(values='passengers',index='month',columns='year',observed=True)
```

```python
sns.heatmap(flight.pivot_table(values='passengers',index='month',columns='year',observed=True))
```

```python
sns.clustermap(flight.pivot_table(values='passengers',index='month',columns='year',observed=True))
```
