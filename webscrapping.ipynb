# Webscrapping
Web scraping (also called web harvesting or web data extraction) is the automated process of extracting information from websites. Instead of copying data manually, a computer program (called a scraper) fetches web pages and then parses (analyzes) the content to collect the desired information.

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

# Accessing The data from website

```python
response=requests.get("https://books.toscrape.com/catalogue/category/books/travel_2/index.html").text
response
```

```python
soup=BeautifulSoup(response,'lxml')
```

```python
print(soup.prettify())
```

```python
for i in soup.find_all('h3'):
  print(i.text)
```

#  Extracting The Books Title

```python
# Extracting the books title
booktitle=[]
for i in soup.find_all('ol',class_='row'):
  for h3_tags in i.find_all('h3'):
    booktitle.append(h3_tags.text)

print(booktitle)
```

# Extracting The Ratings Of The Books

```python
# Extracting the ratings of the books
ratings=[]
for i in soup.find_all('ol',class_='row'):
  for star in i.find_all('p',class_='star-rating'):
    rat=star['class'][1]+' stars'
    ratings.append(rat)
print(ratings)
```

# Extracting The Price Of The Books

```python
# Extracting the price of the books
price=[]
for i in soup.find_all('ol',class_='row'):
  for pr in i.find_all('p',class_='price_color'):
    price.append(pr.text)
print(price)
```

# Extracting the stock availability

```python
# Extracting the stock availability
stock=[]
for i in soup.find_all('ol',class_='row'):
  for st in i.find_all('p',class_='instock availability'):
    stock.append(st.text.strip())
```

# Converting The Data Into The Data Frame(csv file)

```python
data={"Book-Title":booktitle,
      "Ratings": ratings,
      "Price": price,
      "Available": stock}
df=pd.DataFrame(data)
df
```
