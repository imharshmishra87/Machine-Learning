```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',col='sex',row='smoker',hue='sex',kind='line')
```

```python
df=pd.read_csv('/content/placement.csv')
df.head()
```

```python
sns.catplot(data=df,x='iq',hue='placement',kind='violin')
```

```python
coreleation=df.corr()
fig=plt.figure(figsize=(10,10))
sns.heatmap(data=coreleation, annot=True,annot_kws={'size':8},cbar=True,square=True,fmt='.1f',cmap='Blues')
```
